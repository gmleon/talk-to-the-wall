# Participant positions per modality

```python
import pandas as pd
import plotly.express as px
```

```python
hfall = pd.read_csv("data/wall-positions.csv")
hfall.head()
```

```python
hftouch = hfall[hfall["modality"]=="Touch"]
hfspeech = hfall[hfall["modality"]=="Voice"]
```

```python
fontsize=24
```

```python
# Positions in front of the wall while using speech interaction
fig_ht = px.density_heatmap(hfspeech, x="wall_x", y="wall_y", nbinsx=20, color_continuous_scale="Blues", 
                   range_x=[-3000,3000], range_y=[0,4900], width=690, height=400, template="none")

fig_ht.update_xaxes(title="", showticklabels=False)
fig_ht.update_yaxes(title="", tickformat="~s")

fig_ht.update_layout(title="", font_family="Arial", font_size=fontsize, showlegend=False,
    margin=dict(l=70, r=0, b=60, t=15))
fig_ht.layout['coloraxis']['colorbar']['title'] = 'voice'

fig_ht.write_image("fig5-speech.pdf")
fig_ht.show()
```

```python
# Positions in front of the wall while using touch interaction
fig_hs = px.density_heatmap(hftouch, x="wall_x", y="wall_y", nbinsx=20, color_continuous_scale="Greens", 
                   range_x=[-3000,3000], range_y=[0,4900], width=690, height=400, template="none")

fig_hs.update_xaxes(title="", showticklabels=False)
fig_hs.update_yaxes(title="", tickformat="~s")

fig_hs.update_layout(title="", font_family="Arial", font_size=fontsize, showlegend=False,
    margin=dict(l=70, r=0, b=60, t=15))
fig_hs.layout['coloraxis']['colorbar']['title'] = 'touch'

fig_hs.write_image("fig5-speech.pdf")
fig_hs.show()
```
